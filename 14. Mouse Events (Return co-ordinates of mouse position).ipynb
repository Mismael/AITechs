```python
In any Image or Video Capture all what do you want to return in a definite position almost is your point co-ordinates and pixel values.

Each pixel have color values varies by the image type according to the layers and colors of the Image.

RGB: 3 layers (Red, Green, Blue) each one varies from 0-255.
CMYK: 4 layers (Cyan, Magneta, Yellow, Black(key)) almost used in printing systems.
Binary: or white and black image varies between 0 for Black or 1 for White.
Gray Scale: one layer varies from 0-255.
HSV Model: Hue, Saturation, Value or HSV is a color model that describes colors(hue or tint) in terms of their shade (saturation or amount of gray) and their brightness.

and there are multi layers image with very wide colors ranges for accurate colors mostly used in astronomical images.

so now we're going to return the cursor position and pixel color values.
Your result must be like this.....


when you move your cursor ... you will see it's coordinates position and pixel values.

Code

#include <opencv\cv.h>
#include <opencv\highgui.h>
#include <iostream>
#include <opencv2\imgproc\imgproc.hpp>
#include <opencv2\highgui\highgui.hpp>
#include <stdio.h>

IplImage* img;
void mouseEvent(int evt, int x, int y, int flags, void* param){
     evt==CV_EVENT_MOUSEMOVE;
     IplImage* img1=cvCloneImage(img);
     CvScalar s=cvGet2D(img1,y,x);

  unsigned char arr[3];
  arr[0]=s.val[2];
  arr[1]=s.val[1];
  arr[2]=s.val[0];
   
     printf("R=%d, G=%d, B=%d At X=%d Y=%d \n", arr[0], arr[1],arr[2], x, y);
     cvShowImage( "Example1", img1 );
}

int main() 
{
       img = cvLoadImage( "D:\\E-Library\\Programming\\Libraries\\OpenCV\\image.jpg" );
       cvNamedWindow( "Example1", CV_WINDOW_AUTOSIZE );
       cvShowImage( "Example1", img );
       cvSetMouseCallback("Example1",  mouseEvent, 0);
       cvWaitKey(0);
       cvReleaseImage( &img );
       cvDestroyWindow( "Example1" );   
} 

NOTE: if you want to make the same processing on a Camera device frames...just replace all in the main() by the following:

cvNamedWindow( "Example1", CV_WINDOW_AUTOSIZE );
CvCapture*capture =cvCaptureFromCAM(0);
     
  while(1) {
              img = cvQueryFrame( capture );
              if( !img ) break;
             cvShowImage( "Example1", img );
      cvSetMouseCallback("Example1",  mouseEvent, 0);
              char c = cvWaitKey(33);
              if( c == 27 ) break;
}
           cvReleaseCapture( &capture );
           cvDestroyWindow( "Example1" ); 





Explanation:

Briefly, we have image or video/camera capture and cursor.

cvSetMouseCallback(Win name,  mouseEvent, 0): this function is specified to return many parameters specially we concerned with X and Y coordinates which returned by the 2nd parameters, the first parameter is the window name and the third parameter is null.

the second parameter will return the X and Y to "void mouseEvent" and the first parameter in this void will be the mouse event which you can choose one of the following:


CV_EVENT_MOUSEMOVE - when the mouse pointer moves over the specified window
CV_EVENT_LBUTTONDOWN - when the left button of the mouse is pressed on thespecified window
CV_EVENT_RBUTTONDOWN - when the right button of the mouse is pressed on thespecified window
CV_EVENT_MBUTTONDOWN - when the middle button of the mouse is pressed on the specified window
CV_EVENT_LBUTTONUP - when the left button of the mouse is released on thespecified window
CV_EVENT_RBUTTONUP - when the right button of the mouse is released on thespecified window
CV_EVENT_MBUTTONUP - when the middle button of the mouse is released on thespecified window

For the 4th parameter, it will be one of following
CV_EVENT_FLAG_LBUTTON - when the left button of the mouse is pressed on thespecified window
CV_EVENT_FLAG_RBUTTON - when the right button of the mouse is pressed on thespecified window
CV_EVENT_FLAG_MBUTTON - when the middle button of the mouse is pressed on the specified window
CV_EVENT_FLAG_CTRLKEY - when the 'CTRL' key is pressed and any mouse event is occured
CV_EVENT_FLAG_SHIFTKEY - when the 'SHIFT' key is pressed and any mouse event is occured
CV_EVENT_FLAG_ALTKEY - when the 'ALT' key is pressed and any mouse event is occured   

the last parameter in the void takes null value.

and in side the void we will take a copy of the original image by "cvCloneImage".

CvScalar: is a structure container for the values of each layer pixel which returned by "cvGet2D" which takes 3 parameters.... the image and the Y X coordinates values, so to call each value in the container... conatiner name.val[n], at which n is the value of the layer order. ex: 0 for B , 1 for G, 2 for R.
```
