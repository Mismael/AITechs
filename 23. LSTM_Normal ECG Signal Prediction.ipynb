So, the project is divided into 4 Steps <br>
Step 1: Data Loading<br>
Step 2: Split the data into training set and testing set<br>
Step 3: Built a LSTM model to train the data<br>
Step 4: Use the compiled model to predict new data<br>

# Load the used libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers.recurrent import LSTM
from keras.layers.core import Dense, Activation, Dropout
```

# Step 1: Data Loading

```python
data = pd.read_csv('./dataset/Nick-3-EKG_adjusted.txt', sep="\t")
df = data.iloc[:,:8]
data = df["CH1"].values
dataset = np.array(data[:])[:,None]
data = dataset
print("Length of Data", len(data))

plt.figure(figsize=(20,10))
plt.title("Dataset Sample of 3000 Values", fontsize=20)
plt.plot(data[:3000],'r')
```

# Step 2: Split Data into training set and testing set

```python
np.random.seed(1234)
sequence_length = 50
train_start, train_end, test_start, test_end = 0, 3000, 3000, 12000
```

> Training Set Preparation

```python
result = []
for index in range(train_start, train_end - sequence_length):
    result.append(data[index: index + sequence_length])
result = np.array(result)  #shape(samples, sequence_length)
print(result.shape)

plt.figure(figsize=(20,10))
plt.title("Sample of 5 Sequences, Each Sequence of 50 values", fontsize=20)
plt.plot(result[0])
plt.plot(result[1])
plt.plot(result[2])
plt.plot(result[3])
plt.plot(result[4])
```

```python
#z_norm
result_mean = result.mean()
result_std = result.std()
print(result_std)
result -= result_mean
result /= result_std

print ("Mean of train data : ", result_mean)
print ("Train data shape  : ", result.shape)

train = result[:,:]
np.random.shuffle(train)  # shuffles in-place
X_train = train[:, :-1]
y_train = train[:, -1]
#dropin
print("X shape:", X_train.shape)
print("y shape:",y_train.shape)

X_hat = []
y_hat = []
for i in range(0, len(X_train)):
    for j in range(0, np.random.random_integers(0,20)):
        X_hat.append(X_train[i, :])
        y_hat.append(y_train[i])
X_train, y_train = np.asarray(X_hat), np.asarray(y_hat)
print("X shape:", X_train.shape)
print("y shape:",y_train.shape)
```

> Testing Set Preparation

```python
result = []
for index in range(test_start, test_end - sequence_length):
    result.append(data[index: index + sequence_length])
result = np.array(result)  # shape (samples, sequence_length)

plt.figure(figsize=(20,10))
plt.title("Sample of 5 Sequences, Each Sequence of 50 values", fontsize=20)
plt.plot(result[0])
plt.plot(result[1])
plt.plot(result[2])
plt.plot(result[3])
plt.plot(result[4])
```

```python
#z_norm
result_mean = result.mean()
result_std = result.std()
result -= result_mean
result /= result_std

print ("Mean of test data : ", result_mean)
print ("Test data shape  : ", result.shape)

X_test = result[:, :-1]
y_test = result[:, -1]

print("Shape X_train", np.shape(X_train))
print("Shape X_test", np.shape(X_test))

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

# Step 3: Create a LSTM model and compile the data 

```python
model = Sequential()
model.add(LSTM(input_length=sequence_length - 1, input_dim= 1, output_dim= 64, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(output_dim= 1))
model.add(Activation("linear"))
model.compile(loss="mse", optimizer="rmsprop")
    
print("Training")
model.fit(X_train, y_train, batch_size=512, nb_epoch=1, validation_split=0.05)
```

```python
print("Predicting")
predicted = model.predict(X_test)
print(X_test.shape)
print("shape of predicted", np.shape(predicted), "size", predicted.size)
print("Reshaping predicted")
predicted = np.reshape(predicted, (predicted.size,))


plt.figure()
plt.figure(figsize=(15,7))
plt.title("Predicted Signal vs Real Testing Signal", fontsize=20)
plt.plot(predicted[:len(y_test)], 'b', label="Predicted Data")
plt.plot(y_test, 'r', label="Real Testing Data")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.show()
```
