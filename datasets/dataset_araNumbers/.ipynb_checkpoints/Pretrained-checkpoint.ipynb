```python
import numpy as np
from tqdm import tqdm
from glob import glob
from PIL import ImageFile 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_files 

from keras.models import Model , Sequential
from keras.layers import Input , Flatten
from keras import backend as K 
from keras.utils import np_utils
from keras.preprocessing import image
from keras.callbacks import ModelCheckpoint
from keras.applications.inception_v3 import InceptionV3
from keras.layers import Dense, GlobalAveragePooling2D

ImageFile.LOAD_TRUNCATED_IMAGES = True     
```

# Load Dataset

```python
def load_dataset(path):
    data = load_files(path)
    num_files = np.array(data['filenames'])
    num_targets = np_utils.to_categorical(np.array(data['target']), 10)
    return num_files, num_targets
```

```python
'''Change train/test Directories'''
train_files, train_targets = load_dataset('train')
valid_files, valid_targets = load_dataset('test')
num_names = [item[20:-1] for item in sorted(glob("train/*/"))]
```

```python
print('There are %d total Numbers categories.' % len(num_names))
print('There are %s total Numbers images.\n' % len(np.hstack([train_files, valid_files])))
print('There are %d training Numbers images.' % len(train_files))
print('There are %d validation Numbers images.' % len(valid_files))
```

# Preprocessing

```python
def path_to_tensor(img_path): 
    img = image.load_img(img_path, target_size=(84, 84)) 
    x = image.img_to_array(img)
    return np.expand_dims(x, axis=0)

def paths_to_tensor(img_paths):
    list_of_tensors = [path_to_tensor(img_path) for img_path in tqdm(img_paths)]
    return np.vstack(list_of_tensors)
```

```python
train_tensors = paths_to_tensor(train_files).astype('float32')/255
valid_tensors = paths_to_tensor(valid_files).astype('float32')/255
```

```python
plt.imshow(train_tensors[3])
```

# Build Model

```python
input_tensor = Input(shape=(84, 84, 3))  
model0 = InceptionV3(input_tensor=input_tensor, weights='imagenet', include_top=True)

model1 = Sequential()
model1.add(Dense(200 ,input_shape= (1000,), activation='relu'))
model1.add(Dense(10, activation='softmax'))

model = Model(inputs=model0.input, outputs=model1(model0.output))
model.summary()
```

```python
model.compile(optimizer='adam',metrics = ["accuracy"], loss='categorical_crossentropy')
```

```python
epochs = 50

checkpointer = ModelCheckpoint(filepath='saved_models/weights.best.from_scratch.hdf5',verbose=1, save_best_only=True)

model.fit(train_tensors, train_targets, 
          validation_data=(valid_tensors, valid_targets),
          epochs=epochs, batch_size=20, callbacks=[checkpointer], verbose=1)
```
