# 21. Make A Filter

In this Tutorial we will make an intuition step towards the Convolutional Neural Networks (CNN), <br>
CNN mostly concerns about images, even there're some uses in audio signals classification with some edits. but in this tutorial we will focus on images. <br>

Filter (Kernel) is just a matrix window with some values which convolve all over the image to apply some calculation on every  

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bit_me (image_dir, kernel_size):
    image = cv2.imread(image_dir)
    kernel_size = kernel_size
    kernel_step_y = 0
    kernel_step_x = 0
    
    h , w , l = image.shape
    steps_per_width = int(w/kernel_size)
    steps_per_length = int(h/kernel_size)

    new_h = steps_per_length*kernel_size
    new_w = steps_per_width*kernel_size
    image = cv2.resize(image , ( new_w,new_h ))
    
    for j in range (0 , steps_per_length):
        for i in range (0,steps_per_width):
            im = image [0 + kernel_step_y :kernel_size + kernel_step_y , 0 + kernel_step_x :kernel_size + kernel_step_x] #y:y , x:x
            
            im[:,:,0] = int(im[:,:,0].mean()) 
            im[:,:,1] = int(im[:,:,1].mean()) 
            im[:,:,2] = int(im[:,:,2].mean()) 

            image [0 + kernel_step_y :kernel_size + kernel_step_y , 0 + kernel_step_x :kernel_size + kernel_step_x] = im

            kernel_step_x = kernel_step_x + kernel_size
        
        kernel_step_x = 0
        kernel_step_y = kernel_step_y + kernel_size
    
    cv2.imwrite("8_bit_me.jpg",image)    
    image = cv2.cvtColor(image , cv2.COLOR_RGB2BGR)
    plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
    plt.imshow(image)
```

```python
bit_me("PSX_20180402_193219.jpg" , 40)
```
