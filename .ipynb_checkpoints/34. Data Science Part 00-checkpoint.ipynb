```python
#Lib
import numpy as np
import matplotlib.pyplot as plt
import bokeh
import pandas as pd

from bokeh.charts import Bar, output_file, show
from bokeh.io import show, output_notebook
from bokeh.models import LogColorMapper
from bokeh.palettes import Viridis6 as palette
from bokeh.plotting import figure ,output_file
```

```python
dataset_dir = "./datasets/dataset_DS Playground.csv"
```

```python
df = pd.read_csv(dataset_dir)
df.head(5)
```

```python
df = df.groupby(['Gender', 'Exited']).size().unstack(fill_value=0)
```

```python
plt.plot(df)
```

```python
#Count number of values for each Category in Geography Attribute
plt.plot(df['Geography'].value_counts())
```

```python
output_notebook()
p = Bar(df['Geography'].value_counts())
show(p)
```
