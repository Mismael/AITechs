```python
In this post we'll learn some techniques and functions that maybe simple not seem to be important but they are important tools in creating a good project.


Drawing Circle:


cvCircle( image, center, radius, color of circle, thickness, line type, shift);

image: the image you created or you are working on.
center: the coordinates of the center of the circle (X,Y).
radius: radius of the circle...integer value.
color of the circle: the color in RGB values.
thickness: the thickness of the circle, default taken by 1, filled circle taken by negative value (-1).
line type: only one type available for this function (CV_AA).
shift: defaulted by zero.


Drawing Line:
cvLine(image, point1, point2, color of line, thickness, line type, shift);

point1: the start point of the line (X1,Y1).
point2: the end point of the line (X2,Y2).

The rest of the parameters were mentioned in previous function.

Copy an array (image):
cvCopy(array1, array2, 0);

This function copies the first array(image) to another created destination array.

Position of the Created window: 
cvMoveWindow(window, X, Y)

window: the string name of the created window by (cvNamedWindow).
X, Y : the position coordinates.

Create Track Bar:
cvCreateTrackbar( label of the bar, window, initial value, max value, 0);

label of the bar: is the label string written beside the bar like (ex: Value of red:).
initial value: the bar starts with zero value but the initial value is the position of the pointer in the
                     begining of the software.
max value: the maximum value of the bar.

 Create an image window:
 cvCreateImage(size, depth, channels);

size: the size of the created image (width, height).
depth, channels: Click here

Check array range
cvInRangeS(const CvArr* src, CvScalar lower, CvScalar upper, CvArr* dst)

src: first input array.
lowerb: inclusive lower boundary array or a scalar.
upperb: inclusive upper boundary array or a scalar.
dst: output array of the same size as src and CV_8U type.

Compare
cvCmpS(const CvArr* src, double value, CvArr* dst, int cmp_op)

src: single input array.
value: scalar value.
dst: output array that has the same size as the input arrays and type= CV_8UC1 .
cmpop: a flag, that specifies correspondence between the arrays:
                CMP_EQ src1 is equal to src2.
                CMP_GT src1 is greater than src2.
                CMP_GE src1 is greater than or equal to src2.
                CMP_LT src1 is less than src2.
                CMP_LE src1 is less than or equal to src2.
                CMP_NE src1 is unequal to src2.

```
