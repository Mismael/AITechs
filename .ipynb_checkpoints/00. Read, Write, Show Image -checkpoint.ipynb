```python
#Import module
import cv2
import matplotlib.pyplot as plt
```

```python
#read the image from it's directory
image = cv2.imread("/media/mgd-pc/m.magdy.isl/E-Library/Programming/AITechs/img/color-balls.jpg")
```

```python
#to Check the Image shape you can use the following statment 
#it'll return a tuple (image_hight,image_width,number_of_layers)
image.shape
```

```python
#Create Window to show Your Image in
cv2.namedWindow("Image",cv2.WINDOW_AUTOSIZE)
```

```python
#These Cell will Open a Window Contains Your Image
#Finally Show Your Image
cv2.imshow("Image",image)
#Don't Close the window until a key event takes place
cv2.waitKey(0)
#Close the Image Window and Remove it's content from the memory
cv2.destroyWindow("Image")
```

```python
#To Show the Image inside Jupyter

#Opencv Read the Image Layers in BGR sequence,
#so to show it using matplot, we will convert the image to RGB sequence then show it

#Before Converting
plt.figure()
plt.imshow(image)

#After Converting
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB) 
plt.figure()
plt.imshow(image)
```

```python
#Image Resizing,
#cv2.resize(image_variable,(new_width,new_height))
resized_image = cv2.resize(image,(900,500))
```

```python
#Save the new Image
cv2.imwrite("resized.jpg",resized_image)
```
