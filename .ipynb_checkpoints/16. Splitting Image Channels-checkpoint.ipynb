```python
A colored Image consists of 3 main channels.. Red Green Blue (RGB). in wshich each color varies in values from 0-255.

in the next code we'll split a camera captures into 3 channels in 3 frames... Remember that even we're dealing now with a Camera captures (video) but all the process are done on each frame image.

The Code:


int main()
{
cvNamedWindow("rgb");
cvNamedWindow("r");
cvNamedWindow("g");
cvNamedWindow("b");
CvCapture*capture =cvCaptureFromCAM(0);
IplImage* rgb ;
IplImage* r;
IplImage* g;
IplImage* b;

while(1){
rgb= cvQueryFrame( capture );
 r=  cvCreateImage( cvGetSize(rgb), rgb->depth,1 );
 g =  cvCreateImage( cvGetSize(rgb), rgb->depth,1 );
 b =  cvCreateImage(  cvGetSize(rgb), rgb->depth,1 );
cvSplit(rgb,b,g,r,NULL);


cvShowImage("rgb",rgb);
cvShowImage("r",r);
cvShowImage("g",g);
cvShowImage("b",b);

char c = cvWaitKey(33);
if( c == 27 ) break;
}

cvReleaseCapture( &capture );
cvDestroyWindow( "rgb" );
cvDestroyWindow( "r" );
cvDestroyWindow( "g" );
cvDestroyWindow( "b" );

}

Explanation:


our new members in this code are:


cvCreateImage(cvGetSize(FRAME),Depth,Channels): to create a frame image that will hold channel or more of the original one, we must get the size of the original frame. the depth is number of bits per pixel that represents the values of each pixel. 


Depth flags:

IPL_DEPTH_8U - unsigned 8-bit integer. Equivalent to CV_8U in matrix types.
IPL_DEPTH_8S - signed 8-bit integer. Equivalent to CV_8S in matrix types.
IPL_DEPTH_16U - unsigned 16-bit integer. Equivalent to CV_16U in matrix types.
IPL_DEPTH_16S - signed 8-bit integer. Equivalent to CV_16S in matrix types.
IPL_DEPTH_32S - signed 32-bit integer. Equivalent to CV_32S in matrix types.
IPL_DEPTH_32F - single-precision floating-point number. Equivalent to CV_32F in matrix types.
IPL_DEPTH_64F - double-precision floating-point number. Equivalent to CV_64F in matrix types.

OR if you want to use the default of your camera resolution you can just use ("Image name" -> depth).

Channels is the number of channels, in our case we're dealing with 3 channels only but Opencv provide up to 4 channels. Review : http://en.wikipedia.org/wiki/Color_depth

cvSplit(main capture,channel 1,channel 2,channel 3,channel 4): this function splits your main capture into BGR (in order) or CMYK according to your processing... in the case of BGR just write null in the fourth channel. 

```
