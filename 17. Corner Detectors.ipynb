```python
You may find that the title is not such interesting...it's corner detectors !! so ?
but actually it's one of the most processing techniques which you can depend on it in some application like Optical Flow.
but why corners...what is the interesting in it?!
Look at this circle....

Now let me ask you 2 questions....do this circle is rotating ? and where is the most distinct part in the circle?
I know that you can't find the answer....by the way, this is a GIF image for a rotating circle, but do you know why can't you resolve this movement ?! simply...there is no corners to follow.
Optical flow and motion detector applications depends to find the most good features in the frame image and locate corners to follow the motion of objects.

There're two corners detectors algorithms known, Shi-Tomasi and Harris.
Even Shi-Tomasi is more accurate than Harris but we will know how to implement both of them in C language by Opencv Library.

Harris Corner Detector:
as i mentioned previously that we are coding image processing algorithms by a library and to know how they work and there results without get boring with the mathematical calculations.
so let's move to the code and see how it works.

but before that i'd like to inform you that the drawback of harris detector that the corners must be sharp and high defined to be detected so we will use a scale converter which will be used to multiply the pixels values in "scale" value and add "shift" value to enhance the detection, mostly it's weak and not accurate one.

Code:
int main() 
{ 
double minVal=0.0, maxVal=0.0; 
double scale, shift; 
double min=0, max=255; 
  
IplImage * inputImage = cvLoadImage("D:\\image(2).jpg"); 
IplImage *grayImage =cvCreateImage(cvGetSize(inputImage),IPL_DEPTH_8U ,1);

cvCvtColor (inputImage,grayImage,CV_BGR2GRAY);

IplImage *harrisOutput32 =cvCreateImage(cvGetSize(grayImage),IPL_DEPTH_32F ,1); 
IplImage *harrisOutput8 =cvCreateImage(cvGetSize(grayImage),IPL_DEPTH_8U ,1); 

cvCornerHarris(grayImage,harrisOutput32,2); 

cvMinMaxLoc( harrisOutput32, &minVal, &maxVal); 
printf("%d",maxVal);
scale = (max - min)/(maxVal-minVal); 
shift = -minVal * scale + min; 
cvConvertScale( harrisOutput32, harrisOutput8, scale, shift ); 


cvNamedWindow( "Harris", 1); 
cvShowImage( "Harris", harrisOutput8); 
cvWaitKey(0); 
cvReleaseImage (&inputImage); 
cvReleaseImage (&grayImage);   
return 0; 
} 

Before scaling that's what will you see.....


and after.............

as you see that small white dotes denote to the detected corners...so weak detector and can't evaluate the overall corners in the frame...so let's see the next technique.

Shi-Tomasi Detector:

shi-tomasi detector is implemented in this library with a function called "cvGoodFeaturesToTrack".
as we will use this technique later in Optical Flow Application so we need to know how the code work and more definitely, we'll extract these corners by it's co-ordinates as shown below.

Code:

int main (void)
{
 int i, corner_count = 150;

 IplImage* dst_img1 = cvLoadImage ("D:\\image.jpg");
 IplImage* src_img_gray = cvCreateImage (cvGetSize (dst_img1), IPL_DEPTH_8U, 1);
 cvCvtColor(dst_img1,src_img_gray,CV_BGR2GRAY); 
 IplImage* eig_img = cvCreateImage (cvGetSize (src_img_gray), IPL_DEPTH_32F, 1); 
 IplImage* temp_img = cvCreateImage (cvGetSize (src_img_gray), IPL_DEPTH_32F, 1);
 CvPoint2D32f *corners = (CvPoint2D32f *) cvAlloc (corner_count * sizeof (CvPoint2D32f));

 cvGoodFeaturesToTrack (src_img_gray, eig_img, temp_img, corners, &corner_count, 0.1, 15);
 for (i = 0; i < corner_count; i++)
 cvCircle (dst_img1, cvPointFrom32f (corners[i]), 3, CV_RGB (255, 0, 0), 2);

 cvNamedWindow ("Test", CV_WINDOW_AUTOSIZE);
 cvShowImage ("Test", dst_img1);
 cvWaitKey (0);
 cvDestroyWindow ("Test");
 cvReleaseImage (&dst_img1);
 cvReleaseImage (&eig_img);
 cvReleaseImage (&temp_img);
 cvReleaseImage (&src_img_gray);

 return 0;
}


Result:



That's for it know, the code is very simple, but if can't understand it so in the next post we'll use the same code and we'll illustrate it deeply to perform Optical Flow application....Thanks :)  
```
